## Download 4_her spectrum

```python
ssa = 'http://ssaproxy.asu.cas.cz/ccd700/q/pssa/ssap.xml'\
      '?POS={},{}&SIZE=0.05&REQUEST=queryData' \
      '&BAND=6200e-10/6800e-10&format=fits'
her_ra = 238.88
her_dec = 42.55
ssa = ssa.format(her_ra, her_dec)
votable = urllib2.urlopen(ssa).read()
f = StringIO(votable)  # create file for parse_single_table func
table = parse_single_table(f, pedantic=False)
date = (table.array['ssa_dateObs']).data
url = (table.array['accref']).data
name = (table.array['ssa_dstitle']).data
urllib.urlretrieve(url[0], '4_her.fits')
```
